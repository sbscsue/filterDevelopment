```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin,cos,pi
```

```python
plt.rcParams["figure.figsize"] = 10,5
plt.rcParams["font.size"] = 16
```

------

## 1. Make Signal

```python
SR = 500
S = 3
```

```python
#fs
FS = [5,20,100,200]
sig_sub = np.zeros([len(FS),SR*S],float)

for i in range(SR*S):
    for j in range(len(FS)):
        sig_sub[j][i] = sin(2*pi*FS[j]*(i/SR))

```

```python
plt.subplot(3,1,1)
plt.plot(sig_sub[0])
plt.xlim([0,SR])
plt.subplot(3,1,2)
plt.plot(sig_sub[1])
plt.xlim([0,SR])
plt.subplot(3,1,3)
plt.plot(sig_sub[2])
plt.xlim([0,SR])
```

```python
sig = np.zeros(SR*S,float)

for i in range(SR*S):
    for j in range(len(FS)):
        sig[i] += sig_sub[j][i]
```

```python
plt.plot(sig)
```

## 2. FFT

```python
result_real = np.zeros(int((SR/2)))
result_im = np.zeros(int((SR/2)))

for n in range(1,int((SR/2))):
    for x in range(SR*S):
        result_real[n-1] += sig[x] * cos(2*pi*n*(x/SR))
        result_im[n-1] += sig[x] * sin(2*pi*n*(x/SR))
```

```python
plt.plot(result_real)
```

```python
plt.plot(result_im)
```
